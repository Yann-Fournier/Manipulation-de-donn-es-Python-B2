```python
from lxml import etree

# Charger le fichier XML
tree = etree.parse("spatial.xml")

# Définition des xpath
# execution
# affichage

print("Nation name -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/@nom'")
xpath_expression = ".//nation/@nom"
nation_name = tree.xpath(xpath_expression)
print("Resultats:", nation_name, "\n")

print("Nom des astronautes commandants -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission/astronaute[@role='commandant']/text()'")
xpath_expression = ".//nation/programme/mission/astronaute[@role='commandant']/text()"
commandant_name = tree.xpath(xpath_expression)
print("Resultats:", commandant_name, "\n")

print("Date des missions habitées -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[@type='habité']/@date'")
xpath_expression = ".//nation/programme/mission[@type='habité']/@date"
lived_missions_date = tree.xpath(xpath_expression)
print("Resultats:", lived_missions_date, "\n")

print("Mission du premier octobre 1965 -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[@date='1965-01-19'][@type='habité']'")
xpath_expression = ".//nation/programme/mission[@date='1965-01-19'][@type='habité']"
premier_octobre_1965_mission = tree.xpath(xpath_expression)
print("Resultats:", premier_octobre_1965_mission, "\n")

print("Chef mission Apollo 11 -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[@nom='Apollo 11']/astronaute[@role='commandant']/text()'")
xpath_expression = ".//nation/programme/mission[@nom='Apollo 11']/astronaute[@role='commandant']/text()"
chef_mission_Apollo_11 = tree.xpath(xpath_expression)
print("Resultats:", chef_mission_Apollo_11, "\n")

print("Missions sans astronaute -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[count(astronaute) = 0]/@nom'")
xpath_expression = ".//nation/programme/mission[count(astronaute) = 0]/@nom"
missions_sans_astronaute = tree.xpath(xpath_expression)
print("Resultats:", missions_sans_astronaute, "\n")

print("Derniere mission americane du fichier -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation[@nom='USA']/programme[last()]/mission[last()]/@nom'")
xpath_expression = ".//nation[@nom='USA']/programme[last()]/mission[last()]/@nom"
derniere_mission_americane_fichier = tree.xpath(xpath_expression)
print("Resultats:", derniere_mission_americane_fichier, "\n")

print("Mission precedent Voskhod 2 -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[@nom='Voskhod 2']/preceding-sibling::*[1]/@nom'")
xpath_expression = ".//nation/programme/mission[@nom='Voskhod 2']/preceding-sibling::*[1]/@nom"
mission_precedent_Voskhod_2 = tree.xpath(xpath_expression)
print("Resultats:", mission_precedent_Voskhod_2, "\n")

print("Mission un astronaute enregistrer -----------------------------------------------------------------------------------------------")
print("Xpath: './/nation/programme/mission[count(astronaute)=1]/@nom'")
xpath_expression = ".//nation/programme/mission[count(astronaute)=1]/@nom"
mission_un_astronaute_enregistrer = tree.xpath(xpath_expression)
print("Resultats:", mission_un_astronaute_enregistrer, "\n")
```

Partie 3 : Faire des RegEx avec python pour aider Amazone dans la validaition des informations entrées pour la création des comptes

image.png

```python
# Faire une fonction python qui permet de vérifier si les informations d'un individu sont correcre , au moment de la création du compte Amazone:

# - Nom : doit être ne majuscule
# - Prénom : commence apr une majuscule
# - Data de naissance : en jours mois année de deux façons différentes : JJ/MM/AAAA ou JJ Mois AAAA
# - Adresse : Numéro, voie (rue, avenue, ...), ville, pays, code postale

import re

def verification_informations(nom, prenom, date_de_naissance, adresse):
    # Expression régulière pour le nom (doit être en majuscules)
    regex_nom = "^[A-Z]+$"

    # Expression régulière pour le prénom (doit commencer par une majuscule)
    regex_prenom = "^[A-Z][a-z]*$"

    # Expression régulière pour la date de naissance (JJ/MM/AAAA ou JJ Mois AAAA)
    regex_date_naissance = r"^((0[1-9]|[12]\d|3[01])/(0[1-9]|1[0-2])/([0-1]\d{3}|20[0-1][0-9]|202[0-3]))|(0[1-9]|[12]\d|3[01]) (Janvier|Février|Mars|Avril|Mai|Juin|Juillet|Août|Septembre|Octobre|Novembre|Décembre) ([0-1]\d{3}|20[0-1][0-9]|202[0-3])$"

    # Expression régulière pour l'adresse (numéro, voie (rue, avenue, ...), ville, pays, code postal)
    regex_adresse = r"^\d+, (Rue|Avenue|Impasse|rue|avenue|impasse)[\D\s]+, [\D\s]+, [\D\s]+, \d{5}$"

    # Vérification des correspondances
    if re.match(regex_nom, nom):
        print(f"{nom} est un nom valide.")
    else:
        print(f"{nom} n'est pas un nom valide.")

    if re.match(regex_prenom, prenom):
        print(f"{prenom} est un prénom valide.")
    else:
        print(f"{prenom} n'est pas un prénom valide.")

    if re.match(regex_date_naissance, date_de_naissance):
        print(f"{date_de_naissance} est une date de naissance valide.")
    else:
        print(f"{date_de_naissance} n'est pas une date de naissance valide.")

    if re.match(regex_adresse, adresse):
        print(f"{adresse} est une adresse valide.")
    else:
        print(f"{adresse} n'est pas une adresse valide.")

# Exemples de chaînes à vérifier
noms = ["SMITH", "johnson", "JOHNSON", "MacDonald"]
prenoms = ["John", "mary", "Elizabeth", "Robert"]
dates_de_naissance = ["25/12/1990", "01 Janvier 1985", "12/45/2021", "12 Mars 2025"]
adresses = ["123, Avenue Foch, Paris, France, 75012", "quarante_cinq, Impasse du Générale De Gaule, Nanterre, France, 92200", "12, Rue Anatole France, Nanterre, 7RANC3, 12345", "45, Voldemort, Bordeaux, France, 54321"]

for i in range(len(noms)):
    verification_informations(noms[i], prenoms[i], dates_de_naissance[i], adresses[i])
    print("\n", "-----------------------------------------------------------------------------------")
```

```python
# Complétez la fonction de web crawling en faisant une recherche d'hyperliens jusqu'à la 4ème génération (la fonction devra donc être récursive et s'arrêter en fonction d'un critère)
# mémorisez les résultats dans un dictionnaire python sous cette forme :
# web_indexation = { "Famille" : { "parent_1 " : {"parent_1_1" : {
#                                                                    "enfant_1_1_1" : [liste_des_petits_enfants],
#                                                                    "enfant_1_1_2" : petit_enfant
#                                                                },
#                                                },
#                                       "parent_2" : ... ,
#                                        ...
#                                       "parent_n" : ... ,
#                                      }}

import requests
from lxml import html

def crawl_web(url, current_depth=0, max_depth=4, dictionnaire={}):
    if current_depth == max_depth-1: # Cas de base
        response = requests.get(url)
        tree = html.fromstring(response.content)
        links = tree.xpath('//a/@href')
        dictionnaire[url] = links
        return dictionnaire
    
    else: # Cas générale
        response = requests.get(url)
        tree = html.fromstring(response.content)
        links = tree.xpath('//a/@href')
        for link in links:
                if link.strip() and not link.startswith('#'): # Ignorer les liens vides ou les liens d'ancrage
                    if not link.startswith('http'): # Si le lien est relatif, le rendre absolu
                        link = url + link
                    dictionnaire[link] = crawl_web(url=link, current_depth=current_depth+1, max_depth=max_depth, dictionnaire={})
        return dictionnaire

# Exemple d'utilisation
url = "https://google.com"
# url = "https://amazon.com"
# url = "https://www.bienici.com/"
# url = "https://ceweb-agency.com/"
# url = ""
dict_web_crawl = crawl_web(url=url, max_depth=3)
print(dict_web_crawl)
```
